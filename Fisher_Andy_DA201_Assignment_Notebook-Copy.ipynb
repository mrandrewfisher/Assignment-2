**Question 3:** Which month had the highest number of appointments?

```python
# Number of appointments per month == sum of count_of_appointments by month.
# Use the groupby() and sort_values() functions.

ar.groupby('appointment_month')[['count_of_appointments']].agg('sum')\
    .sort_values(by='count_of_appointments', ascending=False)\
    .applymap('{:,.0f}'.format).reset_index()

# ar dataset used as date range is larger
```

**Question 4:** What was the total number of records per month?

```python
# Total number of records per month.

ar_records = ar.groupby('appointment_month')[['count_of_appointments']]\
    .agg('count').sort_values(by='appointment_month').reset_index()

ar_records.columns = ['appointment_month', 'count_of_records']

ar_records
```

<b>Additional query</b> - Investigate missed appointments by location

```python
# generate a ranking of the locations that have the highest % of missed appts

# exclude unknown appointment status, group by location code
ar_loc_known = ar[(ar['appointment_status'] != 'Unknown')].groupby('icb_ons_code')\
    .sum().sort_values(by='count_of_appointments', ascending=False).reset_index()

# rename column names for clarity
ar_loc_known = ar_loc_known.rename(
    columns={'count_of_appointments': 'known appointments'})

ar_loc_dna = ar[(ar['appointment_status'] == 'DNA')].groupby('icb_ons_code')\
    .sum().sort_values(by='count_of_appointments', ascending=False).reset_index()

# rename column names for clarity
ar_loc_dna = ar_loc_dna.rename(
    columns={'count_of_appointments': 'not attended appointments'})

# merge dataframes so not attended % can be calculated
ar_loc_rank = pd.merge(ar_loc_known, ar_loc_dna, on='icb_ons_code')
ar_loc_rank['not attended %'] = ar_loc_rank['known appointments']\
    / ar_loc_rank['not attended appointments']
ar_loc_rank.sort_values(by='not attended %', ascending=False)\
           .head(10).reset_index()
```

```python
# not attended by location, ranked by absolute non attendance
ar_loc_rank.sort_values(by='not attended appointments', ascending=False)\
           .head(10).reset_index()
```

```python
# Create a new DataFrame containing only the text.

tweets_text = tweets['tweet_full_text']

# View the DataFrame.
print(tweets_text)
```

```python
# Loop through the messages, and create a list of values containing the # symbol.

tags = []

for y in [x.split(' ') for x in tweets['tweet_full_text'].values]:
    for z in y:
        if '#' in z:
            # Change to lowercase.
            tags.append(z.lower())

tags
```

### Summary of findings

<b>Network capacity and utilisation of resources needs further investigation<b/>
    
-	Limited conclusions can be drawn regarding utilisation of resources. Capacity limits need to be calculated by practice, knowing staff availability.   
-	A reasonable level of spare capacity should be tolerated so that urgent appointments can be accommodated.

<b>    Weekly seasonality is evident but cannot analyse with missed appointments or utilisation<b/>
-	Knowing that appointment numbers spike on Mondays and drop at weekends suggests that more weekend resource could be required. This should help weekday workers attend appointments without the need for taking time off. However, this may result in higher staffing costs.

<b>    Booking intervals correlate to attendance likelihood<b/>
-	Perhaps there is an overconfidence in patients availability when booking more than 1 day ahead.
-	There appears to be a gradual increasing trend of appointments being missed. Therefore, any remedial actions should be implemented promptly.
-	Due to the data being aggregated into months and lack of clarification in the metadata, it is unclear if the booking and appointment occurred in the same month reported. 

<b>    The cost of different appointment types should be taken into consideration<b/>
-	With the average GP salary range being £65,000-£98,000 it can be assumed that they earn more than most other practice staff, therefore any missed GP appointments will cost the NHS more. Further iterations of analysis should look to apply better cost estimates to missed appointments, rather than the £30 average.
-	It could be assumed that patients are more likely to miss appointments with, say, a physiotherapist, rather than with a GP, as their GP enquiry is likely to be in greater need of diagnosis.

<b>    Missed appointments significantly vary by location<b/>
-	Any practical steps to minimise missed appointments should focus on the locations identified as having the greatest absolute and greatest relative proportion of missed appointments.


### <b> Recommendations<b/>


- Create procedure for data entry standards -> ensures data quality and aids analysis
- Standardise appointment management practices -> if possible, understand the different methods and identify efficiencies or correlation with desired outcome, e.g. costs of missed appointments is beneath acceptable threshold, or % of missed appointments is the lowest.
- Standardise systems -> metadata states multiple issues have been caused by particular systems, indicating a range of systems are used. One robust system would increase resilience and therefore data quality, e.g. standardised inputs/headings, definition of appointment length
- Focus efforts on locations that have the highest absolute and highest percentage of missed appointments.
- Investigate and minimise unknown data. This accounts for a similar number of appointments as those that were missed, so any impact of missed appointments could potentially be up to twice as large.
- Understand patient circumstances before booking so most suitable appointment is selected. e.g. Is this their most convenient practice? Do they need to take time off work to attend? Will they take public transport to the practice? 
- To understand true utilisation of resources, timesheets need to be collected and analysed.
- Understand if courtesy reminders are being sent out and how they could be more effective, e.g. automated.
- Imperial College London showed that fewer appointments would be missed if people were given this information. The NHS could organise Twitter campaigns that could advertise the cost of missed appointments. Subsequent measuring of penetration and traction through scraping retweets and favourites could be plotted against missed appointments to measure effectiveness.
- Investigate the trials of sharing appointments to understand effectiveness.
- Continue to collect the same data and re-use this notebook to monitor changes over time. The code has been designed to be replicable and easy to update for this purpose.
- If possible, covid-related impacts should be sought, quantified and compared with the existing data. A global pandemic will certainly have had an impact, so allowing for this will improve understanding and lead to better results.
