```python
# Load the data set.
tweets = pd.read_csv('tweets.csv')

# View the DataFrame.
tweets.head()
```

```python
# Explore the metadata.

print(tweets.describe())
print(tweets.info())
```

```python
# Explore the data set.

retweet_count = tweets[['tweet_retweet_count']].value_counts().sort_index()

retweet_count
```

```python
favourite_count = tweets[['tweet_favorite_count']].value_counts().sort_index()

favourite_count
```

```python
# Would it be useful to only look at retweeted and favourite tweet messages?
# Explain your answer.

# There are three mechanisms for increasing a tweets visibility or viewcount:
# retweet, favourite, comment on tweet
# There is no functionality button to disapprove of a tweets content or quality.

# Due to it being so easy for anyone to post anything on twitter, retweeted and favourited tweets can be considered a proxy for relevance, validating a tweet as noteworthy 
# however, this mechanism can be abused for nefarious purposes, e.g. inflammatory remarks that are retweeted by twitter bots.

# Placing more emphasis on retweeted and favourite tweet messages is especially useful when they originate from a trusted source.
# Twitter employs a blue tick that lets people know that an account of public interest is authentic.
# Retweets should be considered information deemed worthy of passing on. 
```

```python
# Create a new DataFrame containing only the text.

tweets_text = tweets['tweet_full_text']

# View the DataFrame.
print(tweets_text)
```

```python
# Loop through the messages, and create a list of values containing the # symbol.

tags = []

for y in [x.split(' ') for x in tweets['tweet_full_text'].values]:
    for z in y:
        if '#' in z:
            # Change to lowercase.
            tags.append(z.lower())

tags
```

```python
tags_series = pd.Series(tags)

tags_series_count = tags_series.value_counts()
tags_series_count
```

```python
# Display the first 30 records.

tags_top30 = tags_series_count.iloc[0:30]

tags_top30
```

```python
# Convert the series to a DataFrame in preparation for visualisation.
tags_data = pd.DataFrame(tags_series_count).reset_index()

# Rename the columns.
tags_data.rename(columns={ tags_data.columns[0]: "hashtag", \
                          tags_data.columns[1]: "count"}, inplace = True)
tags_data
```

```python
# Fix the count datatype.
tags_data.dtypes
tags_data['count'].astype(int)

# View the result.
tags_data
```

```python
# Display records where the count is larger than 10.

tags_data_10plus = tags_data[(tags_data['count']>10)]
tags_data_10plus
```

```python
# Create a Seaborn barplot indicating records with a count >10 records.

fig1, ax1 = plt.subplots(figsize=(15,20))
sns.barplot(x='count',
            y='hashtag',
            data=tags_data_10plus)
```

```python
# too many bars make it unclear what to determine from plot
# filter out tweets <20 count to focus on the most relevant tweets
# filter out tweets >100 count as 'healthcare' was the original term searched for 
tags_data_20to100 = tags_data[(tags_data['count']>20) & (tags_data['count']<100)]
fig2, ax2 = plt.subplots(figsize=(15,10))
sns.barplot(x='count',
            y='hashtag',
            data=tags_data_20to100)
```

```python
# Create an aggregated data set to review the different features.
## look at time between appt and booking and if >1 week then understaffed

ar_booking_time = ar.groupby(['time_between_book_and_appointment'])[['count_of_appointments']].sum()

# calculate proportion of when appointments are booked
ar_booking_time['percent'] = ar_booking_time['count_of_appointments']
                            /ar_booking_time['count_of_appointments'].sum()

# View the DataFrame.
ar_booking_time

## appointment status has a possibility of 'available' according to the metadata. 
# However, given that this option does not appear in the dataset, then either:
#     a) there is no spare capacity and therefore staffing levels are insufficient
#     b) available status is removed/amended once the date has been surpassed

# 46% of appointments are for the same day, suggesting spare capacity and therefore spare resource.
# it could be assumed that some of this spare capacity is caused by lack of attendance and also overestimating how long appointments will take
```

```python
# subset of ar looking at non-attendance only

ar_DNA = ar[(ar['appointment_status']=='DNA')]
ar_DNA
```

```python
# groupby booking interval
ar_DNA_booking_time = ar_DNA.groupby(['time_between_book_and_appointment'])\
                    [['count_of_appointments']].sum().reset_index()

# rename columns for clarification
ar_DNA_booking_time = ar_DNA_booking_time.rename(columns={'count_of_appointments':'count_of_missed_appointments'})

# sort values descending so that most unattended appointments are shown first
ar_DNA_booking_time_ordered = ar_DNA_booking_time.sort_values(by='count_of_missed_appointments', ascending=False)
ar_DNA_booking_time_ordered

# plot barplot
sns.barplot(x='time_between_book_and_appointment',
            y='count_of_missed_appointments',
            data=ar_DNA_booking_time_ordered)
```

```python
# Check for outliers
cols = ['count_of_appointments'] 

# Calculate quantiles and IQR.
# Same as np.percentile but maps (0,1) and not (0,100).
Q1 = nc_service_setting[cols].quantile(0.25) 
Q3 = nc_service_setting[cols].quantile(0.75)
IQR = Q3 - Q1
IQR

# Return a Boolean array of the rows with (any) non-outlier column values.
condition = ~((nc_service_setting[cols] < (Q1 - 1.5 * IQR)) \
              | (nc_service_setting[cols] > (Q3 + 1.5 * IQR))).any(axis=1)

# Filter our DataFrame based on condition.
nc_ss_non_outlier = nc_service_setting[condition]
nc_ss_non_outlier.shape

## 11 outliers
```
